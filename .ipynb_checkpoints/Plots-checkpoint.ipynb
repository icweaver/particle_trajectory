```python
%matplotlib inline
from __future__ import division
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import itertools

pi = np.pi
```

### Read in plot data

```python
df_trace  = pd.read_csv('./data/test.dat',delim_whitespace=True) # Trace
df_adj    = pd.read_csv('./data/scratch.dat',delim_whitespace=True) # Trace
df_masses = pd.read_csv('./data/masses.dat',delim_whitespace=True) # Masses
df_impact = pd.read_csv('./data/impact.dat',delim_whitespace=True) # Impact
df_region = pd.read_csv('./data/region.dat',delim_whitespace=True) # Impact region

dummy = df_trace.ix[df_trace['distance'].idxmin()] # Closest approach
closest=str(dummy['distance']) # for closest approach plot text
closest_cm=dummy['distance']*0.03435*10**13 # for closest approach plot text
x_min = dummy['x']
y_min = dummy['y'] 
my_dpi = 96
fig=plt.figure(figsize=(1085/my_dpi, 920/my_dpi), dpi=my_dpi, facecolor='none', edgecolor='none')
ax = fig.add_subplot(111, axisbg='#2E2E2E')
an = np.linspace(0,2*pi,100) # Circularization radius 
ax.plot(0.72*np.cos(an),0.72*np.sin(an), color='lightgreen', linewidth=4.0, mec='lightgreen', label='c.r.')
ax.plot(df_trace['x'],df_trace['y'], 'w', linewidth=2.0, label='tr') # trace
#ax.scatter(df_adj['x'],df_adj['y'],color='pink', marker='o', label='energy loss', s=100) # energy loss

ax.plot(df_masses['x1'],df_masses['y1'], 'yo', ms=180, label='W12', mec='yellow') # WASP 12
ax.scatter(df_masses['x2'],df_masses['y2'],color='orange', marker='o', label='W12b',s=200) # WASP 12b
ax.scatter(df_masses['b1'],df_masses['--'], color='cyan', marker='+' , label='L1',s=600) # L1
#ax.plot([df_masses['x1'],x_min], [df_masses['y1'],y_min], color='#00FF80', linestyle='--', linewidth=2) # closest approach line
#ax.plot(x_min,y_min,'o', label='closest approach', ms=10, mec='pink', color='pink') # closest approach 
#ax.scatter(df_impact['x'], df_impact['y'], label='impact', marker='o', color='red', s=500) # impact
#imp_x = df_impact['x']
#ax.arrow(float(df_impact['x']), float(df_impact['y']), 0.7*float(df_impact['vx']), 0.7*float(df_impact['vy']), 
#         head_width=0.05, head_length=0.1, fc='cyan', ec='cyan', label='impact velocity') # arrow

# Calculating angle 'theta' between velocity vector and position vector 
theta = np.rad2deg(float(df_impact['angle']))
print(theta)

# System
ax.text(-0.99,0.99, 'System: %s' %df_impact['system'].values, color='cyan', fontsize=12)

# Mass ratio
ax.text(-0.99,0.94, 'Mass Ratio: %s' %float(df_impact['mass_ratio']), color='cyan', fontsize=12)

# Radial impact velocity, Keplerian velocity ratio
#ax.text(-0.95,0.80, 'v_R/v_Kep: %s' %(float(df_impact['vr'])/float(df_impact['v_kepler'])),
#color='cyan', fontsize=12)

# Impact velocity
#ax.text(-0.95,0.75, 'Impact Velocity: %s' %float(df_impact['v_total']), color='cyan', fontsize=12)

# Impact Kinetic Energy, Initial Thermal Energy Ratio  
#ax.text(-0.95,0.70, 'K/kT: %s' %(float(df_impact['K_impact'])/float(df_impact['kT_L1'])), 
#color='cyan', fontsize=12)

#ax.text(-0.95, -0.95, 'Closest Approach: %s -->' %closest + ' %0.4E cm' %closest_cm,fontsize=12, color='lightgreen')
#ax.text(-0.95, -0.90, 'Impact Angle: %s' %(theta) + ' degrees', fontsize=12, color='lightgreen')

my_lim = 1.1
plt.xlim([-my_lim, my_lim])
plt.ylim([-my_lim, my_lim])
ax.set_aspect('equal')

ax.set_xlabel('X')
ax.set_ylabel('Y')

for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(14)
    

def flip(items, ncol):
    return itertools.chain(*[items[i::ncol] for i in range(ncol)])

handles, labels = ax.get_legend_handles_labels()

legend = ax.legend(loc=3, numpoints=1, markerscale=0.2, prop={'size':20},
                   scatterpoints=1, ncol=3, frameon=False)
for text in legend.get_texts():
    text.set_color("white")

frame = legend.get_frame()
frame.set_facecolor('#2E2E2E')

#savefig('/Users/KUMKWAT/Desktop/xy.pdf', bbox_inches='tight',dpi=100)
#savefig('/Users/KUMKWAT/Desktop/xy.png', bbox_inches='tight',dpi=100,facecolor=fig.get_facecolor(), edgecolor='none')
#plt.savefig('./Plots/unbound.pdf')
#plt.savefig('/Users/gamma/Desktop/test2.png', dpi=my_dpi, transparent=True)
#plt.savefig('/Users/KIWI/Desktop/artificial_loss.png', bbox_inches='tight',dpi=100,facecolor=fig.get_facecolor(), edgecolor='none')
```

### XYZ Breakdown

```python
fig=plt.figure(figsize=(15, 9), dpi=600, facecolor='none', edgecolor='none')
ax3=fig.add_subplot(111, axisbg='#2E2E2E')

ax3.plot(df_trace['time'],df_trace['x'], label='x', color='cyan', linewidth=2.0)
ax3.plot(df_trace['time'],df_trace['y'], label='y', color='orange', linewidth=2.0)
ax3.plot(df_trace['time'],df_trace['z'], label='z', color='yellow', linewidth=2.0)
ax3.vlines(df_adj['time(binary_period)'],-0.9,0.9, colors='w',linestyles='--', label='lap')

legend = ax3.legend(loc=2)

frame = legend.get_frame()
frame.set_facecolor('lightgrey')

for text in legend.get_texts():
    text.set_color("black")

ax3.set_xlabel('Binary Periods')
ax3.set_ylabel('Displacement (binary separation)')

for item in ([ax3.title, ax3.xaxis.label, ax3.yaxis.label] +
             ax3.get_xticklabels() + ax3.get_yticklabels()):
    item.set_fontsize(14)

#savefig('/Users/ian/Desktop/xyz.pdf',dpi=600)
#savefig('/Users/KUMKWAT/Desktop/xyz.png',dpi=100,facecolor='none',bbox_inches='tight', edgecolor='none')
#savefig('/Users/ian/Desktop/xyz.png', bbox_inches='tight',dpi=100,facecolor=fig.get_facecolor(), edgecolor='none')
```
